```python
import requests
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.flipkart.com/search?q=mobiles%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
content = requests.get(url).text
print(content)
```

```python
soup = BeautifulSoup(content,'lxml')
box = soup.find('div',class_='_1YokD2 _2GoDe3')
```

```python
np = soup.find('a',class_='_1LKTO3').get('href') # next page
print(np)
cnp = 'https://www.flipkart.com/' + np
print(cnp) # combined next page
```

```python
url = cnp
content = requests.get(url).text
soup = BeautifulSoup(content,'lxml')
```

```python
for i in range(2,10):
    url = 'https://www.flipkart.com/search?q=mobiles%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page='+str(i)
    content = requests.get(url).text
    soup = BeautifulSoup(content,'lxml')
    np = soup.find('a',class_='_1LKTO3').get('href') # next page
    cnp = 'https://www.flipkart.com/' + np
    print(cnp)
```

## SCRAPPING DATA

```python
Product_Name = []
Prices = []
Description = []
Reviews = []
Ratings =[]
```

```python
names = box.find_all('div',class_='_4rR01T')
print(names)
```

```python
for i in names:
    name = i.text
    Product_Name.append(name)
print(Product_Name)
```

```python
len(Product_Name)
```

```python
prices = box.find_all('div',class_='_30jeq3 _1_WHN1')
for i in prices:
    price = i.text
    Prices.append(price)
print(Prices)
len(Prices)
```

```python
desc = box.find_all('ul',class_='_1xgFaf')
for i in desc:
    name = i.text
    Description.append(name)
print(Description)
len(Description)
```

```python
rev = box.find_all('div',class_='_3LWZlK')
for i in rev:
    name = i.text
    Reviews.append(name)
print(Reviews)
len(Reviews)
```

```python
ratings = box.find_all('span',class_='_2_R_DZ')
for i in ratings:
    name = i.text
    Ratings.append(ratings)
print(Ratings)
len(Ratings)
```

```python
df = pd.DataFrame({'Product Name': Product_Name,
                   'Prices': Prices,
                   'Description': Description,
                   'Reviews': Reviews})
df
```
